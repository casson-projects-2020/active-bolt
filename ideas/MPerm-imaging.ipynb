<h1>Imagens internas baseadas em relutância magnética</h1>

Trata-se de uma pesquisa a respeito da possibilidade de criar imagens de forma não invasiva da estrutura interna de sólidos utilizando-se relutância magnética.<br/>

O principio é que a maioria dos materiais terá relutância magnética, e essa relutância deve variar por material.

É em teoria possível medir a relutância magnética de um material puro colocando-o dentro de uma interrupção num núcleo de eletromagneto em "C". A intensidade do campo magético deve variar de acordo com a relutância do material colocado na interrupção (inserido dentro da parte do campo magnético que flui através do ar).

A primeira pergunta é: a indutância da bobina que constitui o eletromagneto varia permanentemente (ou apenas temporariamente) proporcionalmente à relutância magnética total do núcleo?

<h2>1 - Dependência entre indutância e núcleo </h2>

A fórmula para a indutância de um circuio magnético é:

(1)\begin{align*}
L = \frac{\Phi B}{I}
\end{align*}

ΦB = link de fluxo magnético, dependente da estrutura geométrica do circuito magnético<br/>

Para o caso específico de indutores lineares: 

(2)\begin{align*}
L = N \frac{\Phi}{I}
\end{align*}

L a indutância em Henries<br/>
N = número de voltas<br/>
Φ = fluxo magnético<br/>
I = corrente <br/>

Sendo que o fluxo magnético depende da corrente, que por sua vez depende da indutância.

O campo magnético no centro de uma espira é:

(3)\begin{align*}
B = \frac{\mu_0 I}{2R}
\end{align*}

μ0 = permeabilidade do espaço vazio = 4π E-7 N/A2<br/>
R = raio da espira<br/>

E o campo magnético no centro de um solenoide com nucleo de ferro:

(4)\begin{align*}
B = \mu \frac{N}{len} I 
\end{align*}

μ = permeabilidade relativa do ferro<br/>
len = comprimento do indutor<br/>

A relação entre fluxo e campo magnetico num solenoide:

(5)\begin{align*}
\Phi = N^2 A B
\end{align*}

A = area da espira

As eqs 2, 4 e 5 permitem chegar em:

(6)\begin{align*}
L = \frac{ N^4 \mu A }{len}
\end{align*}

De forma que parece sim existir uma relação entre a indutância e a permeabilidade (inverso da relutância) do núcleo do indutor.

Para maximimar a influência da relutância é preciso então minimizar a area de espira (o que se adequa ao projeto uma vez que o sensor de relutância deve ser o mais fino possível - uma vez que isso determina a resolução da imagem) e minimizar a quantidade voltas que formam o indutor.

Outro fator para maximizar a influência da variação de relutância é minimizar a relutância do núcleo (uma vez que isso faria com que a relutância do material introduzido na janela de observação - o intervalo aberto no nucleo do sensor - se tornasse mais significativa para o sistema).

Isso indica que a melhor construção seria a de um nucleo de ar - para utilizar a relutância do espaço vazio.



<h2>2 - Relação entre corrente RMS e reatância indutiva</h2>

A fórmula para a reatância indutiva é:

(7)\begin{align*}
X = 2 \pi f L
\end{align*}

O calculo da corrente RMS baseado na reatância é:

(8)\begin{align*}
I_R = \frac{ V_R } X
\end{align*}

De onde decorre imediatamente:

(9)\begin{align*}
I_R = \frac{ V_R } { 2 \pi f L }
\end{align*}


<h2>3 - Comparação entre relutância do ar e de materiais biológicos</h2>

A suscetibilidade magnética dos tecidos vivos varia entre -11 e -7 E-6.

A relação entre permeabilidade do material, do espaço vazio e suscetibilidade é:

(10)\begin{align*}
\mu = \mu_0 (1 + X_m)
\end{align*}

A variação então de permeabilidade entre o ar (de suscetibilidade 0.36 E-6) e os tecidos vivos será:

\begin{align*}
\mu_{air} = \mu_0 (1 + 0.36E-6)
\end{align*}

\begin{align*}
\mu_{min} = \mu_0 (1 -11E-6)
\end{align*}

\begin{align*}
\mu_{max} = \mu_0 (1 -7E-6)
\end{align*}

Nota-se então que a variação de L em (6) será na casa dos 1E-5.

<h2>4 - Principio de funcionamento</h2>

O principio de funcionamento é que os diferentes tecidos humanos (ou, para o caso mais geral, diferentes constituintes de sólidos a obter a imagem interna) terão diferentes relutâncias magnéticas.

A leitura no entanto considera a relutância de uma fatia do material de cada vez. Será então necessário obter um numero de leituras suficiente para resolver o sistema de equações que representa a matriz de pontos que formará a imagem, dentro da resolução desejada.

Por ex, imaginando que a imagem que se deseja tenha uma resolução de 3 x 3 pontos:

Se fizermos 3 leituras na vertical e tres na horizontal, teremos os sistemas de equação:

\begin{align*}A + B + C = val_1\end{align*}
\begin{align*}D + E + F = val_2\end{align*}
\begin{align*}G + H + I = val_3\end{align*}
\begin{align*}A + D + G = val_4\end{align*}
\begin{align*}B + E + H = val_5\end{align*}
\begin{align*}C + F + I = val_6\end{align*}

Nota-se que ainda precisamos de mais três equações para ser capaz de resolver esse sistema.

Podemos obter mais duas leituras nas diagonais:
\begin{align*}A + E + I = val_7\end{align*}
\begin{align*}G + E + C = val_8\end{align*}

E mais equações com menor quantidade de variáveis em outras diagonais:
\begin{align*}D + B = val_9\end{align*}


E em alguns casos, resoluções diretas:
\begin{align*}A = val_{10}\end{align*}

Em teoria pode-se obter o valor diretamente das variaveis A, C, G e I, o que reduziria a complexidade do sistema, mas isso depende do formato do objeto colocado na janela de observação.

Num sistema tridimensional real, no entanto, temos que o sistema não pode presumir que as leituras tenham uma quantidade menor de variaveis, de forma que a eq D + B = val teria que ser D + B + X = val, onde X seria um espaço de ar.

Percebe-se que a quantidade de equações sobe exponencialmente com o aumento de resolução.

Em resoluções maiores no entanto pode-se utilizar a abordagem de girar o sensor em torno do objeto para obter equações em diversos angulos (nesse exemplo de 3 x 3 pontos, isso gera somente 4 equações, sendo que 2 delas em comum com as das leituras horizontal e vertical.

Pode-se também efetuar leituras sequenciais em diversos angulos (desde que a janela de observação seja maior do que o objeto a ser observado) para obter-se uma quantidade suficiente de equações.

<h2>5 - Considerações de projeto</h2>

<h3>A - Minimizar o diametro do indutor que será o sensor</h3> 
Para aumentar a resolução da imagem e para aumentar a influência da variação da relutância magnética com relação à indutância.
<h3>B - Minimizar a quantidade de espiras no indutor</h3> 
Para aumentar a influência da variação da relutância magnética com relação à indutância.
<h3>C - Manter o sensor dentro de uma gaiola de Faraday</h3> 
Já que a variação será na casa de 1E-5, pra evitar contaminação por campos magnéticos externos.
<h3>D - Implementar amostragem (digitalização na leitura)</h3> 
Amostrar o valor analógico, subtrair valor de referência analógico e amplificar em 10.000x a diferença.
<h3>E - Implementar detector de silhueta</h3> 
A silhueta externa permite simplificar os sistemas de equação (impondo que determinados pixels sejam de ar).


<h2>6 - Lista de itens de projeto</h2>

<h3>1 - Sensor em "C"</h3>
Enrolar o indutor em torno de cilindro. Remover o cilindo e moldar a bobina no formato final em "C" (imprimir molde para fazer essa moldagem)
<h3>2 - Gailola de faraday</h3>
Deve ser uma esfera de metal que possa ser aberta, mais a estrutura de suporte para o sensor e para o objeto a ser observado.
<h3>3 - Gerador de sinal senoidal</h3>
Para servir como alimentação para a bobina
<h3>4 - Gerador de valor RMS</h3>
Para linearizar o valor de tensão lida na bobina
<h3>5 - Gerador de tensão de referencia</h3>
Uma tensão variavel de referencia que deve ser subtraída da leitura de tensão do sensor
<h3>6 - Subtrator de tensão</h3>
Utilizar um bom amplificador operacional para subtrair as tensões de leitura e de referencia e amplificar o resultado em 100.000x
<h3>7 - Conversor AD</h3>
Para ler o valor da tensão na saida do subtrator
