```python
import cadquery as cq
from math import sin, cos

s, wa, h = 2.5, 1., 5.5

pl = lambda : cq.Workplane( "XY" )

box = pl().box(*( s,) * 3 )

cbase = pl().box(*( s * 3 + 2 * wa,) * 2, h )
cbase -= pl().box(*( s * 3,) * 2, h ).translate(( 0,0,1))

base = cq.Assembly().add( cbase )
base.add( box.translate(( -s, s, -h/2 + s/2 + wa )))

pos_wp = lambda wp: wp.translate(( -s, 0, -h/2 + s/2 + wa ))

      # peca
cav = pos_wp( box )

# venta
cav += pos_wp( pl().cylinder( 3, .3 )).translate(( 0, 0, 1.5 + s/2 ))

# canal
chan = pos_wp( pl().cylinder( 3, .4 ))
chan = chan.rotate((0,0,0), (0,1,0), 45 )
chan = chan.translate(( s/3 + s * cos( 45 ) ,0, -0.5 ))

cav += chan 

# copo
cav += pl().sphere( 2 ).translate(( 3 * cos( 45 ), 0, 3 ))
cav += pl().cylinder( 2, 2 ).translate(( 3 * cos( 45 ), 0, 5 ))

base.add( cav, color = cq.Color( "purple" ))
display( base )
```
